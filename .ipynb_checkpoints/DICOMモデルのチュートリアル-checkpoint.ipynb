# Voxel画像の準備
## 画像のインポート
* import_dicom()の引数にフルパスを設定します。
* インポートしたモデルは、Objectのvoxcel_model属性に保存されます。

```python
from pymopt.voxel import DicomModel
```

```python
dcm = DicomModel()
dcm.import_dicom('DICOMfile4')
```

## 画像の確認
* Objectのvoxcel_model属性に保存されたDICOM画像データは、display_cross_section()で呼び出せます。
* display_cross_section()は以下表の引数を持ちます。

|  Parameters  |  default  |  type  |    |
| :--- | :--- | :--- | :--- |
|  xx  |  0  |  float  |X軸の断面図位置を設定|
|  yy  |  0  |  float  |y軸の断面図位置を設定|
|  zz  |  0  |  float  |z軸の断面図位置を設定|
|  int_pix  |  True  |  bool  |  int_pix = False で mmを入力|
|  cmap  |  'gray'  |  matplotlib.cmap  |画像のカラーマップを設定|
|  graph_type  |  'ALL'  |  {'ALL','XY','XY-YZ','NO_HIST'}  |グラフの出力数を設定|



```python
dcm.display_cross_section(xx = 500,yy = 500,zz = 500,cmap = 'gray')
```

### 計測面の確認
* 本課題での計測面は、以下のようにOリングが設置している方向から光を照射します。<br>
つまり、Z = 0 mm の基準面に設定する必要があります。<br>
ちなみに、Oリングの穴は、計測試料設置時にの上下を表しており、穴の空いている方が上になります。

```python
dcm.display_cross_section(xx = 500,yy = 500,zz = 1020,graph_type = 'XY-YZ',cmap = 'gray')
```

* rot180_y()は、y軸を基準に３次元画像を180度回転させます。<br>このメソッドは、voxel_model属性を変更します。<br>
もし画像を元に戻したい場合、もう一度 rot180_xz()を使います。
* Z軸方向基準で回転させたい場合、rot90_z()メソッドを使います。<br>
これは、Numpy.rot90と同じで、引数はkは何回90度回転させるかを設定します。

```python
dcm.rot180_y()
dcm.display_cross_section(xx = 500,yy = 500,zz = 80,graph_type = 'ALL',cmap = 'gray')
```

## トリミング
* この状態のままだと、Oリングが邪魔で計算ができません。そのため、不要な部分をトリミングします。<br>
また、トリミングは、モデルのデータサイズを削減することが可能です。
* トリミングでは、まずトリムする領域をtrim_area()メソッドで確定します。
* trim_area()メソッドは以下の引数を持ちます。

|  Parameters  |  default  |  type  |    |
| :--- | :--- | :--- | :--- |
|  right  |  0  |  float  |X軸でX最大値から負方向にトリムする長さ|
|  left  |  0  |  float  |X軸でX最小値から正方向にトリムする長さ|
|  upper  |  0  |  float  |Y軸でY最大値から負方向にトリムする長さ|
|  lower  |  0  |  float  |Y軸でY最大値から正方向にトリムする長さ|
|  top  |  0  |  float  |Z軸でZ最小値から正方向にトリムする長さ|
|  bottom  |  0  |  float  |Z軸でZ最大値から負方向にトリムする長さ|
|  int_pix  |  True  |  bool  |  int_pix = False で mmを入力|
|  cmap  |  'gray'  |  matplotlib.cmap  |画像のカラーマップを設定|
|  Graph_type  |  'ALL'  |  {'ALL','XY','XY-YZ','NO_HIST'}  |グラフの出力数を設定|

```python
trim_params = {
    'int_pix':False,
    'right': 0, 
    'left': 0,
    'upper': 0,
    'lower':0,
    'top':4.8,
    'bottom':0,
    'cmap':'PiYG',
}
dcm.trim_area(**trim_params)
```

左上に表示される画像は正面画像で、色が若干薄くなっているのが、トリミングされる範囲です。<br>次に、実際に画像をトリミングします。


### トリミング
* トリミングは、set_trim()メソッドで行います。トリミングされるのは、trim_area()で設定した領域になります。
* このメソッドでは、トリミングする前に、閾値以下の値を削除します。<br>
閾値を設定したい場合は、thresholdの引数で値を設定してください。引数の値は、画像のデータタイプで異なるので、ヒストグラムの値を参考にしてください。また、予め値を確認したい場合、reConstArray()で確認できます。何も入れない場合、thresholdはデフォルト値が参照されます[Watanabe 2018]。
* トリミングを間違ってしまった場合<br>
この時は、reset_setting()メソットで画像を再読み込みする形で設定をリセットできます。しかしながら、全ての設定がリセットされるので、初めからやった方が早いかもしれません。

```python
dcm.set_trim(cmap = 'gray')
```

# Voxelモデルのモデリング
## モデルの肉付

* 画像を編集した後、モンテカルロで用いるデータをモデリングします。<br>最初は、海綿骨のみのモデルに皮膚や緻密骨などを追加していきます。

## 光学特性の付与
* それぞれの組織は、波長に対して独自の光学特性を有しています。ここでは、吸収係数、散乱係数、異方性係数、屈折率の値を設定します。
* ここでは、２値モデルと線形モデルで異なる値の設定方法を行う必要があります。
* 光学特性は、set_params(**params) メソッドで設定します。
